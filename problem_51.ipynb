```python
def prime(a):
    if a<2:
        return False
    elif a==2:
        return True
    else:
        for i in range(2,int(a**0.5)+1):
            if a%i==0:
                return False
        return True
def permute(elements, current):
    if len(elements) == 0: 
        return [current] 
    else:
        result = []
        for i in range(len(elements)):
            remaining = elements[:i] + elements[i+1:]
            new_current = current + elements[i]
            result.extend(permute(remaining, new_current))
        return result

def list_calculator(d):
    if d==0:
        return [""]
    number_list=list()
    number=1
    while True:
        if len(str(number))==d:
            number_list.append(number)
        elif len(str(number))>d:
            break
        number+=1
    return number_list


def prime_family(a):
    
    a=str(a)
    for i in range(len(str(a))):
        prime_family_list=list()
        left_a=a[:i]
        right_a=a[i+1:]
        list_1=list()
        for x in list_calculator(len(left_a)):
            list_1.append(str(x)+a[i])
        list_2=list()
        for y in list_calculator(len(right_a)):
            list_2.append(str(y))
        for x in list_1:
            for y in list_2:
                new_a=x+y
                if prime(int(new_a)):
                    prime_family_list.append(new_a)
        if len(prime_family_list)==8:
            prime_family_list.sort()
            print(prime_family_list[0])
            return True
        print(prime_family_list)
    return False
number=2
while True:
    if prime(number) and prime_family(number):
        break
    number+=1
    print(number)
```
